## DataFrame - 그룹분석

```python
import numpy as np
import pandas as np
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head(3)
```

```python
# 품종별, sepla_length의 평균
iris.groupby('species')['sepal_length'].mean()
```

```python
# 그룹핑 후 열 선택 
setosa = iris[iris.species == 'setosa']
setosa
```

```python
# 필요한 열 먼저 선택하고 그룹 분석
iris[['sepal_length', 'species']].groupby('species').mean()
```

```python
iris.groupby('species').mean()
```

```python
# 모든 피처에 대한 품종별 평균, 표준편차 
res = iris.groupby('species').agg(['mean','std'])
res
```

```python
# res는 컬럼인덱스가 두개, 중복해서 
res['sepal_length']['mean']['setosa']
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

- guessing
성별, 흡연유무, 점심/저녁

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.head()
```

```python
# 성별에 따른 tip_pct의 평균
tips.groupby('sex')[['tip_pct']].mean()
```

```python
# 흡연유무, 시간에 따른 tip_pct의 평균
tips.groupby(['smoker', 'time'])[['tip_pct']].mean()
```

```python
# 세부류중 어떤 비율이 가장 팁의 비율이 높았는지
res = tips.groupby(['sex','smoker','time'])[['tip_pct']].mean()
res
```
