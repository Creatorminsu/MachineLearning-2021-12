Anscombe's Quartet

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ans = sns.load_dataset("anscombe")
ans.head(3)
```

```python
ans.shape
```

* dataset 1 에 대해서 그래프를 그려봄

```python
ds = ans.plot()
```

```python
ds = ans[ans.dataset == 'I']
ds
```

- 회귀식 y = 0.5 * x + 3

```python
plt.scatter(ds.x, ds.y)
plt.plot([4,14], [5, 18], 'r-', lw=2)
plt.title('Dataset I')
plt.show()
```

* 4개의 그래프를 하나의 그림에 그리기

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, roman in enumerate(['I','II','III','IV']):
    ax = axs[i//2 , i % 2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20], [4.5,13], 'r-',lw=2)
    ax.set_ylim(2,13)
    ax.set_title(f"Dataset {roman}")
```

* 최종그림

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8), sharex=True, sharey=True)
for i, roman in enumerate(['I','II','III','IV']):
    ax = axs[i//2 , i % 2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20], [4.5,13], 'r-',lw=2)
    ax.set_ylim(2,13)
    ax.set_title(f"Dataset {roman}")
plt.suptitle("Anscombe's Quartet", fontsize=20)
plt.tight_layout()
plt.show()
```
